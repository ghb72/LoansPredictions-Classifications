Lets see the mutual information score

```python
import sklearn
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression
from sklearn.preprocessing import LabelEncoder
```

```python
#EmploymentStatus	EducationLevel	MaritalStatus	HomeOwnershipStatus	LoanPurpose
preprocessed_data = loans.copy()
le = LabelEncoder()
for col in preprocessed_data.select_dtypes('object').columns:
    preprocessed_data[col] = le.fit_transform(preprocessed_data[col])
    # Mostrar la asignacin de categoras a nmeros
    label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print(label_mapping)
```

For a Classification of Approved or not in the loans we can see the mutual information

```python
target = preprocessed_data['LoanApproved']
preprocessed_data = preprocessed_data.infer_objects()

m_infos = []
cols = []
for col in loans.columns:
    if col != 'LoanApproved':# and col != 'ApplicationDate':
        m_info = mutual_info_classif(preprocessed_data[[col]], target)
        m_infos.append(m_info[0])
        cols.append(col)

plot = sns.barplot(x=m_infos, y=cols, orient='h')
# plot = sns.barplot(x=cols, y=m_infos)
# ticks = plt.xticks(rotation=90)
title = plt.title('Mutual Information between variabels and Loan Approval')
```

We have the predictor and non predictor features for classify in approved or non aproved Loan

```python
variables = pd.DataFrame(index=cols, data={'variable':m_infos},)
to_prop_cols_classify = variables[variables['variable']==0].index.values
print('Non predictor Feature:', variables[variables['variable']==0].index.values)
print('Predictor Feature:', variables[variables['variable']>0].index.values)
```

## Clean and preprocess data

```python
import numpy as np
import scipy.stats as stats
## we dont have missed values
def outliers_calc(df, method = 'z-score', threshold = 3):
    # 3 sigmas of deviation threshold
    outliers_id = []
    outliers_values = []
    if method == 'z-score': # we have less errors with z-score
        for i, v in enumerate(df):
            mean = df.mean()
            std = df.std()
            t = (v-mean)/std
            if t > threshold:
                outliers_id.append(i)
                outliers_values.append(v)
            else:
                continue
    elif method == 'mad':
        median = df.median()
        mad = np.abs(stats.median_abs_deviation(df))
        for i, v in enumerate(df):
            t = (v-median)/mad
            if t > threshold:
                outliers_id.append(i)
                outliers_values.append(v)
            else:
                continue
    return outliers_id, outliers_values
```

The count of outliers by Z-score method

```python
rows_to_drop = []
for col in preprocessed_data.columns:
    outliers_id, outliers_values = outliers_calc(preprocessed_data[col], method='z-score')
    rows_to_drop=rows_to_drop +outliers_id
    print(f'For {col} we have {len(outliers_id)} outliers')
```

```python
index_to_drop = list(set(rows_to_drop))
print(len(index_to_drop)) # Count of rows with outliers
```

```python
preprocessed = preprocessed_data.drop(index_to_drop)
preprocessed.head()
```
